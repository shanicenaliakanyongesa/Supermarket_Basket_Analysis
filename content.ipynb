```python
import pandas
data = pandas.read_csv("https://modcom.co.ke/data/datasets/shared_articles.csv")
data
```

```python
data.info()
```

```python
# We will only work with text colm

X = data['text'].values
X

#X
```

```python
from sentence_transformers import SentenceTransformer#It is a library which transforms this sentenses to vectors
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')#This is the model that will transform our sentense to vectors
embeddings = model.encode(X, show_progress_bar = True)#here we actually apply the model to our sentenses and show the progress(Changing X to the actual vector) Takes the sentense ang generates an encoding for each one.
embeddings
```

```python
df = pandas.DataFrame(embeddings)
df
# machinelearningplus.com/nlp/cosine-similarity
```

```python

from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(df)
# 1 the two vectors are identical in direction.
# A  0 indicates  meaning there is no similarity.
# A -1 indicates that the vectors are diametrically opposed.
```

```python
# pass above to be like a dataframe
cosine_data = pandas.DataFrame(similarity)
cosine_data
```

```python
#Find similar indexes 
#Find similar articles
#We pass the id of the article we want to find similar articles
#cosine_data.loc[newsid]: This selects the row in cosine_data corresponding to newsid. This row contains the cosine similarity scores of the given article with all other articles.
#sort.values-most similar articles come first.
#After sorting, we convert the indices of the sorted values into a list. We then take the indices from the second to the tenth ([1:10]).
#4. Retrieving Article Titles

def recommendations(newsid):#id ya articles
    index_recomendations=cosine_data.loc[newsid].sort_values(ascending=False).index.tolist()[1:10]#Finding Similar Articles
    print("Similar Indexes", index_recomendations)
    news_titles=data['title'].loc[index_recomendations].values
    print("Similar Articles", news_titles)
recommendations(3)#ths is the id

#The recommendations function finds articles similar to a given article ID (newsid) using cosine similarity scores stored in cosine_data.
# It retrieves the indices of the most similar articles (excluding the article itself) and then retrieves and prints the titles of these similar articles from the data DataFrame.
```
